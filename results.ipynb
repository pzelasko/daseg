# Parse new journal experiments

```python
def is_result_ok(path):
    # discarded result - erroneous run?
    path = str(path)
    if all(tok in path for tok in ('xlnet', 'turn')):
        print('discarded:', path)
        return False
    return True
    

root = Path('/Users/pzelasko/jhu/da/08-23-2020-results/')
pkls = [p for p in root.rglob('results.pkl') if is_result_ok(p)]
# add fixed turns results
pkls += [p for p in Path('/Users/pzelasko/jhu/da/08-28-2020-results-xlnet-turns-fixed/').rglob('results.pkl')]
pkls_noprop = list(root.rglob('results_noprop.pkl'))
pkls_lf_a1_l4096 = list(Path('/Users/pzelasko/jhu/da/08-25-2020-results-nogradacc/').rglob('results.pkl'))
pkls_lf_a1_l512 = list(Path('/Users/pzelasko/jhu/da/08-26-2020-results-nogradacc-seq512/').rglob('results.pkl'))
sorted(pkls + pkls_noprop + pkls_lf_a1_l4096 + pkls_lf_a1_l512)
```

```python
def retrieve(data):
    return {**data['sklearn_metrics'], **data['zhao_kawahara_metrics']}

def parse(name: str, note: str) -> dict:
    ps = name.split('_')
    return {
        'model': ps[0],
        'dataset': ps[1],
        'context': ps[2],
        'case': ps[3],
        'tagset': ps[4],
        'seed': ps[5],
        'note': note
    }

def determine_note(name: str) -> str:
    if 'dialog' in name:
        if 'xlnet' in name:
            return 'prop'
        return 'l4096_a6'
    return ''
        
def read_bigru(root):
    import re
    bigru_logs = sorted(p for p in root.rglob('stdout.txt') if 'bigru' in str(p))
    pattern = re.compile(r'TEST RESULTS\n([^-]+)\'test_loss\'', re.MULTILINE)
    for p in bigru_logs:
        logs = p.read_text()
        match = pattern.search(logs)
        if match is None:
            print("could not read:", p)
            continue
        results_text = match.group(1) + '}' if match is not None else None
        parsed_text = {k: v for k, v in eval(results_text).items() if not k.startswith('ZK_')}
        yield {**parse(p.parent.name, note=''), **parsed_text}
        

results = {p.parent.name: retrieve(pickle.load(open(p, 'rb'))) for p in pkls if 'bigru' not in str(p)}
flat_results = [
    {**parse(name, note=determine_note(name)), **metrics} for name, metrics in results.items()
]
flat_results += [{**parse(p.parent.name, note='noprop'), **retrieve(pickle.load(open(p, 'rb')))} for p in pkls_noprop if 'bigru' not in str(p)]
flat_results += [{**parse(p.parent.name, note='l4096_a1'), **retrieve(pickle.load(open(p, 'rb')))} for p in pkls_lf_a1_l4096]
flat_results += [{**parse(p.parent.name, note='l512_a1'), **retrieve(pickle.load(open(p, 'rb')))} for p in pkls_lf_a1_l512]
flat_results += list(read_bigru(root))

with open('full_results.pkl', 'wb') as f:
    pickle.dump(results, f)
with open('flat_results.pkl', 'wb') as f:
    pickle.dump(flat_results, f)
#results.update({p.parent.name: pickle.load(open(p, 'rb')) for p in pkls if 'bigru' in str(p)})
```

```python
pd.set_option('precision', 2)
df = (
    pd.DataFrame(flat_results)
    .drop(['micro_f1_common_I', 'macro_f1_common_I'], axis=1)
    .sort_values(by=['case', 'dataset', 'tagset', 'context', 'model', 'note', 'seed'])
    .rename({'SegmentationWER': 'SegWER'}, axis=1)
)
df.head()

mean_df = df.groupby(['case', 'dataset', 'tagset', 'context', 'model', 'note']).mean() * 100
std_df = df.groupby(['case', 'dataset', 'tagset', 'context', 'model', 'note']).std() * 100
mean_df
```

```python
from itertools import groupby

def list_tb(root: str, apply_filter=False):
    paths = sorted(p for p in Path(root).rglob('events.out.tfevents.*') if not apply_filter or is_result_ok(p))
    for key, group in groupby(paths, key=lambda p: p.parent.parent.parent.name):
        latest_version_path = sorted((p.parent.name.replace('version_', ''), p) for p in group)[-1][1]
        yield latest_version_path

def parse_tb_file(path, note=None):
    name = path.parent.parent.parent.name
    props = parse(name, note=note if note is not None else determine_note(name))
    try:
        ea = event_accumulator.EventAccumulator(str(path))
        ea.Reload()
        #print(ea.Tags())
        val_losses = ea.Scalars('val_loss')
        df = (
            pd.DataFrame(ea.Scalars('val_loss'))
            .rename({'value': 'val_loss'}, axis=1)
            .drop('wall_time', axis=1)
            .assign(
                epoch=list(range(1, len(val_losses) + 1)),
                macro_f1=pd.DataFrame(ea.Scalars('macro_f1')).value,
                **props
            )
        )
        df = df.drop(df[df.epoch > 10].index)
        return df
    except:
        print('Couldnt parse:', path)
        #raise
        
    

all_dfs = []
for tb_file in list_tb('/Users/pzelasko/jhu/da/08-23-2020-results', apply_filter=True):
    all_dfs.append(parse_tb_file(tb_file))
for tb_file in list_tb('/Users/pzelasko/jhu/da/08-28-2020-results-xlnet-turns-fixed/'):
    all_dfs.append(parse_tb_file(tb_file))
for tb_file in list_tb('/Users/pzelasko/jhu/da/08-25-2020-results-nogradacc/'):
    all_dfs.append(parse_tb_file(tb_file, note='l4096_a1'))
for tb_file in list_tb('/Users/pzelasko/jhu/da/08-26-2020-results-nogradacc-seq512/'):
    all_dfs.append(parse_tb_file(tb_file, note='l512_a1'))
df = pd.concat(all_dfs)
df
```

```python
from daseg import DialogActCorpus, Call, FunctionalSegment
from typing import Set, Tuple
from collections import defaultdict
from cytoolz import sliding_window
from itertools import chain
```

```python
swda = DialogActCorpus.from_swda_path('deps/swda/swda', splits=['test'])
```

```python
def find_context(start_word_idx, end_word_idx, call):
    print(start_word_idx, end_word_idx)
    wc = 0
    for seg, next_seg in sliding_window(2, chain(call, [None])):
        n_words = seg.num_words + int(next_seg is not None and seg.speaker != next_seg.speaker)
        bounds = (wc, wc + n_words)
        #print(f'{bounds[0]} <= {start_word_idx} <= {bounds[1]}')
        if bounds[0] <= start_word_idx <= bounds[1]:
            yield seg
        wc += n_words

def find_error_segments(true_dataset: DialogActCorpus, pred_dataset: DialogActCorpus):
    counts = {
        'DSER': 0,
        'SegmentationWER': 0,
        'DER': 0,
        'JointWER': 0,
    }
    total_segments = 0
    total_words = 0

    def build_segment_set(call: Call, with_labels: bool) -> Set[Tuple]:
        segment_set = set()
        word_pos = 0
        for segment in call:
            n_words = len(segment.text.split())
            segment_set.add((word_pos, word_pos + n_words) + ((segment.dialog_act,) if with_labels else ()))
            word_pos += n_words
        return segment_set
    
    errors = defaultdict(lambda: defaultdict(list))

    try:
        for cid, true_call, pred_call in zip(true_dataset.call_ids, true_dataset.calls, pred_dataset.calls):
            true_segments = build_segment_set(true_call, with_labels=False)
            pred_segments = build_segment_set(pred_call, with_labels=False)
            error_segments = true_segments - pred_segments
            for seg_idx, segment in enumerate(error_segments):
                errors[cid]['seg'].append({
                    'ref': list(find_context(segment[0], segment[1], true_call)),
                    'hyp': list(find_context(segment[0], segment[1], pred_call))
                })
                counts['DSER'] += 1
                counts['SegmentationWER'] += segment[1] - segment[0]
                raise 1

            true_segments = build_segment_set(true_call, with_labels=True)
            pred_segments = build_segment_set(pred_call, with_labels=True)
            error_segments = true_segments - pred_segments
            for seg_idx, segment in enumerate(error_segments):
                errors[cid]['joint'].append(true_call[seg_idx])
                counts['DER'] += 1
                counts['JointWER'] += segment[1] - segment[0]

            total_segments += len(true_call)
            total_words += len(true_call.words(add_turn_token=False))
    except:
        pass

    return errors
```
